# WBK - Análise de Empréstimos e Inadimplência

A WBK  **é uma empresa fictícia** especializada em empréstimos, com foco na facilidade e agilidade para seus clientes. Todo o processo de solicitação e gestão de empréstimos é otimizado para ser realizado de forma rápida e prática via dispositivos móveis, oferecendo a melhor experiência ao usuário.

O objetivo da empresa é fornecer o maior número possível de informações sobre os empréstimos, garantindo total transparência para seus clientes. No entanto, a inadimplência tem sido uma grande preocupação para a WBK. Atualmente, a taxa de inadimplência está em 35%, e a empresa busca implementar técnicas de análise para reduzir esse percentual para 25% ou menos.

Para isso, fui responsável pela análise dos dados históricos de empréstimos, fornecidos pela empresa. O banco de dados contém informações detalhadas sobre os últimos empréstimos concedidos, e com base nesses dados, a análise de inadimplência será realizada, utilizando técnicas de modelagem preditiva e análise estatística.

# Importando bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sqlalchemy import create_engine  # pip install SQLAlchemy psycopg2 (Instala o SQLAlchemy caso ainda não tenha)
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

# Conectando-se com o banco de dados SQL

```python
# Faz a conexão com o banco de dados
engine = create_engine('postgresql://postgres:1234567@localhost/credito')

# Comando sql para puxar os dados da tabela CREDITO
sql = 'SELECT * FROM "CREDITO";'

# Leitura dos dados sql com a biblioteca pandas
df = pd.read_sql(sql, engine)
```

# Etapa de análise exploratória

```python
# Verificando o Dtype das colunas e consequentemente verificando se há nulos
df.info()
```

```python
# Verificando as 5 primeiras linhas do dataframe
df.head()
```

```python
# A coluna 'Profissao' tem uma grande quantidade de informações desnecessárias, que podem atrapalhar no aprendizado de máquina
df['Profissao']
```

```python
# Remoção das colunas 'IDCREDITO' e 'Profissao'
df.drop(['IDCREDITO', 'Profissao'], axis=1, inplace=True)
```

```python
# Os dados nulos pertencem as colunas 'Emprego' , 'ResidenciaDesde' e 'Habitacao'
df.isnull().sum()
```

## Tratamento dos Dados

```python
# Foram trocados os dados nulos pelas medianas de cada uma
```

```python
median = df['Emprego'].median()
```

```python
df.loc[df['Emprego'].isnull(),'Emprego'] = median
```

```python
median2 = df['ResidenciaDesde'].median()
```

```python
df.loc[df['ResidenciaDesde'].isnull(),'ResidenciaDesde'] = median2
```

```python
median3 = df['Habitacao'].median()
```

```python
df.loc[df['Habitacao'].isnull(),'Habitacao'] = median3
```

```python
# Verificando se ainda há algum dado nulo
df.isnull().sum()
```

# Separando os dados para treinamento e teste

```python
# Dessa maneira se obtêm todos os dados excluindo o dado alvo (neste caso o dado alvo é a coluna 'Status')
x = df.drop('Status',axis=1)
```

```python
# Pega apenas os dados da coluna alvo
y = df['Status']
```

```python
# Etapa de separação de dados para treino e teste
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

# Etapa de treinamento dos modelos de Machine Learning

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
lr = LogisticRegression(max_iter=6000)
lr.fit(x_train,y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

# Etapa para ver quais modelos tem os melhores Scores

## Score do RandomForestClassifier

```python
rfScore = rf.score(x_test,y_test)
print("Score de {:.2f}%".format(rfScore*100))
```

## Score do LogisticRegression

```python
lrScore = lr.score(x_test,y_test)
print("Score de {:.2f}%".format(lrScore*100))
```

## score do GaussianNB

```python
gnbScore = gnb.score(x_test,y_test)
print("Score de {:.2f}%".format(gnbScore*100))
```

## Score do KNeighborsClassifier

```python
knnScore = knn.score(x_test,y_test)
print("Score de {:.2f}%".format(knnScore*100))
```

## Score da DecisionTreeClassifier

```python
dtScore = dt.score(x_test, y_test)
print("Score de {:.2f}%".format(dtScore*100))
```

# Conclusão

**Exemplo ilustrativo**<br>

Imagine que você tenha 100 clientes, sendo 35 inadimplentes e 65 adimplentes (taxa de inadimplência de 35%).

O modelo tem acurácia de ~80%.
Suponha que, dentre os 35 inadimplentes, o modelo identifique corretamente 28 (TP = 28) e erre 7 (FN = 7). Isso dá um Recall de 80% (28/35).
Dentre os 65 adimplentes, digamos que o modelo classifique corretamente 52 (TN = 52) e erre 13 (FP = 13).

Se você recusa todos os que o modelo diz serem inadimplentes:<br>**Recusados:** TP (28 inadimplentes) + FP (13 adimplentes) = 41 pessoas.<br>
**Aprovados:** FN (7 inadimplentes) + TN (52 adimplentes) = 59 pessoas.

Entre os aprovados, só 7 são inadimplentes (FN).

A taxa de inadimplência na carteira aprovada passa a ser 7/59 ≈ **11.86%**.

Nesse cenário hipotético, a taxa de inadimplência cai bem **abaixo de 25%**. Porém, você também recusou **13 bons pagadores (FP)**.

>***Se baseando no exemplo acima temos as seguintes taxas de inadimplência a seguir:***

```python
inadRF = (cm1[1][0] / (cm1[1][0] + cm1[1][1]))*100
print("Taxa de inadimplência com o RandomForestClassifier ≈ {:.2f}%".format(inadRF))
```

```python
inadLR = (cm2[1][0] / (cm2[1][0] + cm2[1][1]))*100
print("Taxa de inadimplência com a LogisticRegression ≈ {:.2f}%".format(inadLR))
```

```python
inadGNB = (cm3[1][0] / (cm3[1][0] + cm3[1][1]))*100
print("Taxa de inadimplência com o GaussianNB ≈ {:.2f}%".format(inadGNB))
```
